1. What is Simple Linear Regression (SLR)? Explain its purpose.
- Simple Linear Regression (SLR) is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.

- It tries to find the best-fitting line that predicts the value of Y based on X.

- Purpose

  - To predict values

  - To understand relationships between variables

  - To estimate trends

- Example: Predicting house price based on size.
2. What are the key assumptions of Simple Linear Regression?
- Simple Linear Regression works properly only when certain assumptions are satisfied:

- Linearity: The relationship between X and Y must be linear.

- Independence: Observations should be independent of each other.

- Homoscedasticity: The variance of errors should remain constant.

- Normality of Errors: The residuals (errors) should be normally distributed.

- No Outliers: Extreme values should not heavily affect the model.
3. Write the mathematical equation for a simple linear regression model and explain each term.
- y = mx + b
- Explanation of each term:
| Term  | Meaning                                       |
| ----- | --------------------------------------------- |
| **y** | Dependent variable (value we want to predict) |
| **x** | Independent variable (input feature)          |
| **m** | Slope of the line (rate of change)            |
| **b** | Intercept (value of y when x = 0)             |

- The equation represents the best fit line for the data.

5. What is the method of least squares in linear regression?
- The Least Squares Method is used to find the best-fitting line.
- It works by minimizing the total squared difference between actual and predicted values.
- Formula concept:
  Error = (Actual - Predicted)^2

- The regression line is chosen such that: Sum of all squared errors is minimum.
- Why square the errors?
  - Prevents positive and negative errors from canceling out
  - Gives more weight to large errors

6. What is Logistic Regression? How does it differ from Linear Regression?
- Logistic Regression:Logistic Regression is used for classification problems where the output is categorical.

- Example:

  - Spam or Not Spam

  - Pass or Fail

  - Disease or No Disease

- It predicts probability between 0 and 1.
| Feature | Linear Regression   | Logistic Regression       |
| ------- | ------------------- | ------------------------- |
| Output  | Continuous value    | Categorical / probability |
| Example | Predict salary      | Predict pass/fail         |
| Graph   | Straight line       | S-shaped curve            |
| Use     | Regression problems | Classification problems   |

10. How do you interpret the coefficients in a simple linear regression model?
- The coefficient (slope) tells how much the dependent variable changes when the independent variable increases by 1 unit.
- Example

- Regression equation:
Salary=2000+500×Experience
- Interpretation:

  - Intercept (2000) → Starting salary when experience = 0

  - Slope (500) → Salary increases by 500 units for each year of experience

- So if experience increases by 1 year, salary increases by 500.

```
#9 Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 5, 4, 5])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
